```python
import pygame
import random
pygame.init()

#Colors
white=(255,255,255)
red=(255,0,0)
black=(0,0,0)

gameWindow=pygame.display.set_mode((900,600))

#game title
pygame.display.set_caption("SnakeGame")
pygame.display.update()

#Game Variables
exit_game=False
game_over=False
snake_x=115
snake_y=55
velocity_x=0
velocity_y=0
score=0
food_x=random.randint(450,900)
food_y=random.randint(300,600)
snake_size=10
food_size=snake_size
fps=30#frames per second

clock=pygame.time.Clock()

while(not exit_game):
    for event in pygame.event.get():
        print(event)
    if(event.type==pygame.KEYDOWN):
        if(event.key==pygame.K_RIGHT):
            snake_x=snake_x+10
            velocity_x=10
        if(event.key==pygame.K_LEFT):
            snake_x=snake_x-10
            velocity_x=10
        if(event.key==pygame.K_UP):
            snake_y=snake_y-10
            velocity_y=10
        if(event.key==pygame.K_DOWN):
            snake_y=snake_y+10
            velocity_y=10
    if(abs(snake_x-food_x)<6and abs(snake_y-food_y)<6):
       score=score+10
       snake_size+=5
       print("score=",score)
       food_x=random.randint(0,900)
       food_y=random.randint(0,600)
       
    
            
    gameWindow.fill(white)
    pygame.draw.rect(gameWindow,black,[snake_x,snake_y,snake_size,snake_size])
    pygame.draw.rect(gameWindow,red,[food_x,food_y,food_size,food_size])
    pygame.display.update()
    clock.tick(fps)
```
